```python
def cleanup_csv(df_):
    df = df_.dropna(subset=['Area Id', 'Variable Id', 'Year', 'Value'], how='all').copy()
    df['Area Id'] = df['Area Id'].astype(int)
    df['Variable Id'] = df['Variable Id'].astype(int)
    df['Year'] = df['Year'].astype(int)
    return df
```

```python
source_dir = '../source/new'
```

```python
all_data = []

for f in os.listdir(source_dir):
    if 'csv' in f:
        path = os.path.join(source_dir, f)
        try:
            df = cleanup_csv(pd.read_csv(path, index_col=False))
            all_data.append(df)
        except:
            print(f)
```

```python
all_concepts = []

for df in all_data:
    all_concepts.append(df[['Variable Name', 'Variable Id']].copy())
```

```python
all_concepts = pd.concat(all_concepts)
```

```python
all_concepts = all_concepts.reset_index(drop=True).drop_duplicates()
```

```python
all_concepts['concept'] = all_concepts['Variable Name'].map(lambda x: to_concept_id(x))

all_concepts.columns = ['name', 'variable_id', 'concept']
```

```python
all_concepts['concept_type'] = 'measure'
```

```python
all_area = []

for df in all_data:
    all_area.append(df[['Area', 'Area Id']].copy())
```

```python
all_area = pd.concat(all_area)
```

```python
all_area = all_area.drop_duplicates()
```

```python
all_area['area'] = all_area['Area'].map(lambda x: x.split('|')[0])
all_area['name'] = all_area['Area'].map(lambda x: x.split('|')[1])
```

```python
all_area = all_area.drop('Area', axis=1)
```

```python
all_area['area'] = all_area['area'].map(to_concept_id)
```

```python
all_area.columns = ['area_id', 'area', 'name']
```

```python
all_area
```

```python
all_area[['area', 'name', 'area_id']].to_csv('../../ddf--entities--area.csv', index=False)
```

```python
for df in all_data:
    assert len(df['Variable Name'].unique()) == 1, 'more than one indicator in this dataframe'   
    concept = to_concept_id(df.loc[0, 'Variable Name'])
    if concept in cdf_dup.index:
        var_id = df.loc[0, 'Variable Id']
        if var_id != cdf_dup.loc[concept, 'variable_id']:
            print(f'dropping duplicated data for {concept}, id={var_id}')
            continue
    df = df.rename(columns={'Value': concept, 'Year': 'year'})
    df['area'] = df['Area'].map(lambda x: to_concept_id(x.split('|')[0]))    
    df[['area', 'year', concept]].to_csv('../../ddf--datapoints--{}--by--area--year.csv'.format(concept), index=False)
```
